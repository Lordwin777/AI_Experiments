```python
import pandas as pd
import numpy as np
from sklearn.metrics.pairwise import cosine_similarity
```

```python
import pandas as pd

# Assuming the dataset is extracted to '/content/'

# Load the ratings data (u.data)
r_cols = ['userId', 'movieId', 'rating', 'timestamp']
ratings_100k = pd.read_csv('/content/u.data', sep='\t', names=r_cols, encoding='latin-1')

# Load the movie data (u.item)
m_cols = ['movieId', 'title', 'release_date', 'video_release_date', 'imdb_url', 'unknown', 'Action', 'Adventure', 'Animation', 'Children\'s', 'Comedy', 'Crime', 'Documentary', 'Drama', 'Fantasy', 'Film-Noir', 'Horror', 'Musical', 'Mystery', 'Romance', 'Sci-Fi', 'Thriller', 'War', 'Western']
movies_100k = pd.read_csv('/content/u.item', sep='|', names=m_cols, encoding='latin-1')

# Display the first few rows of the loaded data
print("MovieLens 100K Ratings Data (u.data):")
display(ratings_100k.head())
print("\nMovieLens 100K Movie Data (u.item):")
display(movies_100k.head())
```

```python
# Merge ratings and movie data to get movie titles for recommendations
merged_data_100k = pd.merge(ratings_100k, movies_100k[['movieId', 'title']], on='movieId', how='left')
display(merged_data_100k.head())
```

```python
# Create the user-item matrix for MovieLens 100K dataset
user_item_100k = merged_data_100k.pivot_table(
    index='userId',
    columns='movieId',
    values='rating'
).fillna(0)

display(user_item_100k.head())
```

```python
# Calculate user similarity for the MovieLens 100K dataset
user_similarity_100k = cosine_similarity(user_item_100k)
user_similarity_df_100k = pd.DataFrame(
    user_similarity_100k,
    index=user_item_100k.index,
    columns=user_item_100k.index
)

display(user_similarity_df_100k.head())
```

```python
import pandas as pd
import numpy as np

# Helper function to get movie titles from movie IDs
def get_movie_titles(movie_ids, movies_df):
    return movies_df[movies_df['movieId'].isin(movie_ids)]['title'].tolist()

def recommend_movies(user_id, user_item_matrix, user_similarity_df, top_n_recommendations=10):
    # Get the similarity scores for the target user
    user_similarity = user_similarity_df.loc[user_id]


    similar_users = user_similarity.drop(user_id).sort_values(ascending=False)

    # Get the movies rated by the target user
    user_rated_movies = user_item_matrix.loc[user_id]
    # Get the movies that the target user has not rated
    unrated_movies = user_rated_movies[user_rated_movies == 0].index

    # Dictionary to store weighted sum of ratings and sum of similarities for unrated movies
    movie_scores = {}

    # Iterate through similar users
    for sim_user_id, similarity_score in similar_users.items():
        if similarity_score > 0: # Only consider users with positive similarity
            # Get movies rated by the similar user
            sim_user_ratings = user_item_matrix.loc[sim_user_id]

            # Filter for movies that the target user has not rated but the similar user has
            relevant_movies = sim_user_ratings[sim_user_ratings > 0].index.intersection(unrated_movies)

            for movie_id in relevant_movies:
                if movie_id not in movie_scores:
                    movie_scores[movie_id] = {'weighted_sum': 0, 'similarity_sum': 0}

                # Add weighted rating to the movie score
                movie_scores[movie_id]['weighted_sum'] += sim_user_ratings[movie_id] * similarity_score
                movie_scores[movie_id]['similarity_sum'] += similarity_score

    recommendations = {}
    for movie_id, scores in movie_scores.items():
        if scores['similarity_sum'] > 0:
            recommendations[movie_id] = scores['weighted_sum'] / scores['similarity_sum']

    # Convert to a Pandas Series, sort and get top N recommendations
    recommendations_series = pd.Series(recommendations)
    recommendations_series = recommendations_series.sort_values(ascending=False)

    return recommendations_series.head(top_n_recommendations)

# Example: Recommend movies for a specific user (e.g., userId = 1)
user_id_100k = 1
top_n_recommendations_100k = 10

recommendations_100k = recommend_movies(
    user_id_100k,
    user_item_100k,
    user_similarity_df_100k,
    top_n_recommendations=top_n_recommendations_100k
)

recommended_movie_ids = recommendations_100k.index.tolist()
recommended_movie_titles = get_movie_titles(recommended_movie_ids, movies_100k)

print(f"\nTop {top_n_recommendations_100k} movie recommendations for User {user_id_100k} (MovieLens 100K Dataset):")
for i, title in enumerate(recommended_movie_titles):
    print(f"{i+1}. {title}")
```
