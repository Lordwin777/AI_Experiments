```python
import numpy as np
import networkx as nx
import re
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.metrics.pairwise import cosine_similarity
```

```python
text = """My name is Lordwin Joseph, and I am currently pursuing B.Tech in ICT.
I have a strong interest in Cyber Security and Cloud Computing, where I enjoy exploring how systems can be secured and optimized.
I have worked on projects such as a Web Scraper and PhishLens, which helped me understand real-world applications of security and data handling.
I am also deeply interested in exploring AI, web technologies, and cybersecurity, as these areas continue to evolve and offer new challenges.
One of my key strengths is my ability to understand concepts clearly and apply them effectively though it may take time.
My goal is to build a successful career in cybersecurity and secure a job in this field."""
```

```python
sentences = re.split(r'(?<=[.!?]) +', text)
```

```python
def preprocess(sentence):
    sentence = re.sub(r'[^a-zA-Z0-9 ]', '', sentence)
    return sentence.lower()
```

```python
clean_sentences = [preprocess(s) for s in sentences]
```

```python
vectorizer = TfidfVectorizer()
tfidf_matrix = vectorizer.fit_transform(clean_sentences)
```

```python
sim_matrix = cosine_similarity(tfidf_matrix, tfidf_matrix)
```

```python
nx_graph = nx.from_numpy_array(sim_matrix)
```

```python
scores = nx.pagerank(nx_graph)
```

```python
ranked_sentences = sorted(((scores[i], s) for i, s in enumerate(sentences)), reverse=True)
```

```python
def get_summary(n):
    return " ".join([ranked_sentences[i][1] for i in range(n)])
```

```python
# 50% summary (approx)
print("50% Summary:\n")
print(get_summary(len(sentences)//2))
```

```python
print("\n25% Summary:\n")
print(get_summary(max(1, len(sentences)//4)))
```
