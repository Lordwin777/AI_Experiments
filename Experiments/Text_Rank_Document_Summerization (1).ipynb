```python
def preprocess_sentence(sentence):
  stop_words = set(stopwords.words('english'))
  words = word_tokenize(sentence.lower())
  words = [w for w in words if w.isalnum() and w not in stop_words]
  return words
```

```python
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.metrics.pairwise import cosine_similarity

def sentence_similarity(sent1, sent2):
  # Preprocess sentences using the existing function
  words1 = preprocess_sentence(sent1)
  words2 = preprocess_sentence(sent2)

  # If either sentence becomes empty after preprocessing, similarity is 0
  if not words1 or not words2:
    return 0.0

  # Join words back into strings for TF-IDF Vectorizer
  processed_sent1 = ' '.join(words1)
  processed_sent2 = ' '.join(words2)

  # Create TF-IDF vectorizer and fit it to both sentences
  vectorizer = TfidfVectorizer()
  tfidf_matrix = vectorizer.fit_transform([processed_sent1, processed_sent2])

  # Calculate cosine similarity
  similarity = cosine_similarity(tfidf_matrix[0:1], tfidf_matrix[1:2])[0][0]
  return similarity
```

```python
def build_sim_mat(sentences):
  n = len(sentences)
  sim_mat = np.zeros((n, n)) # Corrected: passed shape as a tuple
  for i in range(n):
    for j in range(n):
      if i != j:
        sim_mat[i][j] = sentence_similarity(sentences[i],sentences[j])
  return sim_mat
```

```python
def textrank_summarization(text, num_sentences = 3):
  sentences = sent_tokenize(text)
  sim_mat = build_sim_mat(sentences)
  graph = nx.from_numpy_array(sim_mat)
  scores = nx.pagerank(graph)
  ranked_sentences = sorted(((scores[i],s) for i,s in enumerate(sentences)), reverse = True)
  summary = [sentence for score, sentence in ranked_sentences[:num_sentences]]
  return ' '.join(summary)
```
