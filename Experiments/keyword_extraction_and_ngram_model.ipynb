```python
def textrank_keyword(text,topn=5,window_size=4):
  words=word_tokenize(text.lower())
  stop_words=set(stopwords.words('english'))
  words=[w for w in words if w not in stop_words and w not in string.punctuation]
  graph = nx.Graph()

  for i, word in enumerate(words):
    for j in range(i+1, i+window_size+i):
      if(j<len(words)):
        graph.add_edge(word,words[j])

  scores = nx.pagerank(graph)
  ranked_words=sorted(scores.items(),key=lambda x:x[1],reverse=True)
  keywords = [word for word, score in ranked_words[:topn]]
  return keywords
```

```python
print(textrank_keyword(text))
```

## **N-GRAM MODEL WORD PREDICTION**

```python
corpus = """
I like machine learning
I like deep learning
I enjoy machine learning
I enjoy data science
Machine learning is interesting
Deep learning is powerful
"""
```

```python
import nltk
from nltk.util import ngrams
from collections import defaultdict, Counter
import random
```

```python
nltk.download('punkt')
nltk.download('punkt_tab')
```

```python
tokens = nltk.word_tokenize(corpus.lower())
print(tokens)
```

```python
# build n grams model
n = 2
bigrams = list(ngrams(tokens, n))
print(bigrams)
```

```python
#create the probability model
model = defaultdict(Counter)
for w1,w2 in bigrams:
  model[w1][w2]+=1

model
```

```python
# function to predict the next word
def predict_next_word(word):
  word = word.lower()
  if word in model:
    next_words = model[word.lower()]
    prediction=max(next_words,key=next_words.get)
    return prediction
  else:
    return "Word not found in model"
```
